```python
#importing libraries 
import spacy
 
# Load the English language model
nlp = spacy.load("en_core_web_sm")
```

```python
content = []

with open("/OTTO-Project/FE_Documents_Txt/EMMA_Official_Statement/ER988258-ER773536-ER1174834.txt", encoding="utf8") as f:
    content = f.readlines()
```

```python
content
```

```python

for line in content:
    if "cusip" in line.lower():
        # Process the text with SpaCy
        doc = nlp(line)
        # Display the PoS tagged result
        # print("Original Text: ", line)
        # print("PoS Tagging Result:")
        pos_list = [token.pos_ for token in doc]
        dom_pos = max(set(pos_list), key = pos_list.count)
        if dom_pos == "PROPN":
            print(line)
        # for token in doc:
        #     print(f"{token.pos_}", sep=" ")
```
